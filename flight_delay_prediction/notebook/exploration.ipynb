#### Handle the missing values

```python
# Drop rows where SCHEDULED_TIME is null
df.dropna(subset = ["SCHEDULED_TIME"])
```

```python
# List of delay-related columns
delay_columns = ['AIR_SYSTEM_DELAY', 'SECURITY_DELAY', 'AIRLINE_DELAY', 'LATE_AIRCRAFT_DELAY', 'WEATHER_DELAY']

# Impute missing values with 0 in these columns
df[delay_columns] = df[delay_columns].fillna(0)
```

```python
# Fill missing values in CANCELLATION_REASON with "Not Cancelled"
df['CANCELLATION_REASON'] = df['CANCELLATION_REASON'].fillna('Not Cancelled')
```

```python
# Impute missing TAIL_NUMBER values with the mode (most frequent value)
df['TAIL_NUMBER'].fillna(df['TAIL_NUMBER'].mode()[0], inplace=True)
```
