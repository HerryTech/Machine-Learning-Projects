#### Import necessary libraries

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.pipeline import make_pipeline
```

#### Load and Preprocess Data

#### Split Data

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, stratify = y, random_state = 1)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

#### Model Training

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

#### Model Evaluation

```python
train_pred = model.predict(X_train)
train_accuracy = accuracy_score(train_pred, y_train)
```

```python
train_accuracy
```

```python
test_pred = model.predict(X_test)
test_accuracy = accuracy_score(test_pred, y_test)
```

```python
test_accuracy
```

#### Grid Search

```python
model_params= {
    "Logistic_Regression": {
        "model": LogisticRegression(),
        "params": {
            "logisticregression__C": [0.001, 0.01, 0.1, 1, 10, 100],
            "logisticregression__solver": ["liblinear", "saga", "lbfgs"],
        }
    },
    
    "Svm": {
        "model": SVC(),
        "params": {
            "svc__C": [0.1, 1, 10, 100],
            "svc__kernel": ['linear', 'rbf', 'poly']
        }
    },
    
    "Decision_Tree": {
        "model": DecisionTreeClassifier(),
        "params": {
            'decisiontreeclassifier__criterion': ['gini', 'entropy'],
            'decisiontreeclassifier__max_depth': [3, 5, 10, 20, None],
            'decisiontreeclassifier__min_samples_split': [2, 5, 10, 20],
        }
    },
    
    "Random_Forest": {
        "model": RandomForestClassifier(),
        "params": {
            'randomforestclassifier__n_estimators': [50, 100, 200, 500],
            'randomforestclassifier__max_depth': [10, 20, 30, None],
            'randomforestclassifier__min_samples_split': [2, 5, 10, 20],
            'randomforestclassifier__min_samples_leaf': [1, 2, 5, 10],
        }
    }
}
```

```python
score = []
best_estimators = {}
for algo, mp in model_params.items():
    pipe = make_pipeline(StandardScaler(), mp["model"])
    clf = GridSearchCV(pipe, mp["params"], cv = 5, n_jobs = -1)
    clf.fit(X_train, y_train)
    score.append({
        "model": algo,
        "best_score": clf.best_score_,
        "best_params": clf.best_params_
    })
    best_estimators[algo] = clf.best_estimator_
```

```python
df = pd.DataFrame(score)
```

```python
best_estimators["Svm"].score(X_test, y_test)
```

```python
best_estimators["Logistic_Regression"].score(X_test, y_test)
```

```python
best_estimators["Decision_Tree"].score(X_test, y_test)
```

```python
best_estimators["Random_Forest"].score(X_test, y_test)
```

#### Best Model Evaluation

```python
best_model = best_estimators["Svm"]
```

```python
best_pred = best_model.predict(X_test)
```

```python
clf_accuracy = accuracy_score(y_test, best_pred)
clf_accuracy
```

```python
cm = confusion_matrix(y_test, best_pred)
cm
```
